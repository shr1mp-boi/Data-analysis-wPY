## **2. Operaciones Básicas con Series**
#### - Suma dos series de números.
#### - Filtra una `Serie` para mostrar solo los valores mayores a 5.
#### - Calcula la media, mediana y desviación estándar de una `Serie`

```python
ser1 = pd.Series(np.arange(1,6))
print("Serie 1:\n", ser1)
ser2 = pd.Series(np.arange(6,11))
print("Serie 2:\n", ser2)
```

```python
# SUMA DE DOS SERIES DE NUM

series_sum = ser1 + ser2
series_sum
```

```python
# FILTRAR UNA SERIE

np.random.seed(123)
ser3 = pd.Series(np.random.randint(1, 15, size = 15))
print("Serie antes de filtrar data:\n", ser3)

sermask = (ser3 > 5)
print("\nSerie después de filtrar data:\n")
ser3[sermask]
```

```python
# CALCULAR MEAN, MEDIAN y STD de una Serie
np.random.seed(123)
ser4 = pd.Series(np.random.randint(1,20, size = 20))
ser4
```

```python
#PROMEDIO
ser4.mean()
```

```python
# MEDIANA 
ser4.median()
```

```python
# DESVIACION ESTANDAR
ser4.std()
```

## **3. DataFrames Básicos**
#### - Crea un `DataFrame` con nombres de alumnos y sus calificaciones en dos exámenes.
#### - Agrega una columna que calcule el promedio de las calificaciones.

```python
df = pd.DataFrame({'Alumnos': ['Gael', 'Luz', 'Octavio', 'Daniela', 'Juan'],
    'Calificacion 1': [9, 8, 6, 7, 10], 'Calificacion 2': [7, 9, 6, 10, 8]})
df
```

```python
df['Promedio'] = df[['Calificacion 1', 'Calificacion 2']].mean(axis = 1)
df
```

# **Ejercicios Intermedios con Funciones**

## **4. Uso de Funciones**
#### - Aplica una función `lambda` para aumentar en un 5% las calificaciones.
#### - Crea una función que convierta las calificaciones a letras (A, B, C, etc.) y aplícala al `DataFrame`

```python
df = (
    df.assign(
        calif1_5percent = lambda x: x['Calificacion 1'] + x['Calificacion 1']*0.05,
        calif2_5percent = lambda x: x['Calificacion 2'] + x['Calificacion 2']*0.05
    )
) 
df
```

```python

df = (
    df.assign(
        Calif1_letras = lambda x: x['Calificacion 1'].apply(
            lambda nota: "A+" if "10" in str(nota) else 
                                "A" if "9" in str(nota) else 
                                "B" if "8" in str(nota) else 
                                "C" if "7" in str(nota) else
                                "D"
        ),

        Calif2_letras = lambda x: x['Calificacion 2'].apply(
            lambda nota: "A+" if "10" in str(nota) else 
                                "A" if "9" in str(nota) else 
                                "B" if "8" in str(nota) else 
                                "C" if "7" in str(nota) else
                                "D"
        )
    )
)

df
```
